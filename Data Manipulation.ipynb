```python
newCols = {
    'timestamp' : 'timestamp',
    'cnt':'count',
    't1' : 'realTemp_C',
    't2' : 'feelsLikeTemp_C',
    'hum' : 'humidity_percent',
    'wind_speed' : 'wind_speed_kph',
    'weather_code' : 'weather',
    'is_holiday' : 'is_holiday' ,
    'is_weekend' : 'is_weekend',
    'season' : 'season'           
}

bikes_df.rename(newCols, axis = 1, inplace = True)
```

```python
bikes_df.head()
```

```python
def categorize_season(season):
    if   season == 0.0: return 'spring'
    elif season == 1.0: return 'summer'
    elif season == 2.0: return 'autumn'
    elif season == 3.0: return 'winter'

bikes_df['season'] = bikes_df['season'].apply(categorize_season)
```

```python
bikes_df.head()
```

```python
def categorize_weather(weather):
    if   weather == 1.0: return 'Clear'
    elif weather == 2.0: return 'Scattered clouds'
    elif weather == 3.0: return 'Broken clouds'
    elif weather == 4.0: return 'Cloudy'
    elif weather == 7.0: return 'Rain'
    elif weather == 10.0: return 'Rain with thunderstorm'
    elif weather == 26.0: return 'Snowfall'

bikes_df['weather'] = bikes_df['weather'].apply(categorize_weather)
```

```python
bikes_df.head()
```

```python
bikes_df['humidity_percent'] = (bikes_df['humidity_percent']/100)
```
